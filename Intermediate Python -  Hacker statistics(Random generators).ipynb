**Empire State game simulation:**

```python
# Dice simulation:

seed = np.random.seed(123)
total_steps = 0
outcomes = []

for x in range(1, 101):
    for index in range(1, 101):
        dice_number = np.random.randint(1, 7)
        print(f'{index} dice number: {dice_number}')

    if dice_number < 3:
        total_steps = total_steps - 1
        outcomes.append(total_steps)
    elif np.logical_or(dice_number >= 3, dice_number < 6):
        total_steps = total_steps + 1
        outcomes.append(total_steps)
    else:
        roll_dice_again = np.random.randint(1, 7)
        total_steps = total_steps + roll_dice_again
        outcomes.append(total_steps)
print(f'You have walked {total_steps} steps!')
print(f'Outcomes: {outcomes}')
    
```
